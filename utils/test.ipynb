```python
def scan_dir_rename(root_path, old_word='modificated', new_word='modified'):
    print(f"Scanning: {root_path}")
    lista = os.listdir(root_path)
    for l in lista:
        actual_path = os.path.join(root_path, l)
        if os.path.isdir(actual_path):
            scan_dir_rename(actual_path)
            if 'modificated' in l:
                os.rename(actual_path, actual_path.replace(old_word, new_word))
                print(f"RENAMED:  {l}")
        if os.path.isfile(actual_path) and 'modificated' in l:
            os.rename(actual_path, actual_path.replace(old_word, new_word))
            print(f"RENAMED:  {l}")

scan_dir_rename('./')
```
