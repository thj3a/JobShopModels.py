```python
import os
import pandas as pd
```

```python
folder_path = './results/csv/timestamp/'
filenames = os.listdir(folder_path)
for filename in filenames:
    path = os.path.join(folder_path, filename)
    df = pd.read_csv(path, sep=';')

    # Check if there is a job that starts before the previous one finishes in the same machine
    machines = df['Resource'].unique()
    for machine in machines:
        df_machine = df[df['Resource'] == machine]
        df_machine = df_machine.sort_values(by=['Start'])
        df_machine = df_machine.reset_index(drop=True)
        for i in range(len(df_machine)-1):
            if df_machine.loc[i, 'Finish'] > df_machine.loc[i+1, 'Start']:
                print('Error in file: ' + filename + ' on machine: ' + str(machine))
                print(df_machine.loc[i, 'Finish'], df_machine.loc[i+1, 'Start'])
    
    # Check if there is a job that starts before the previous one finishes in different machines
    jobs = df['Job'].unique()
    for job in jobs:
        df_job = df[df['Job'] == job]
        df_job = df_job.sort_values(by=['Op'])
        df_job = df_job.reset_index(drop=True)
        for i in range(len(df_job)-1):
            if df_job.loc[i, 'Finish'] > df_job.loc[i+1, 'Start']:
                print('Error in file: ' + filename + ' on job: ' + str(job))
                print(df_job.loc[i, 'Finish'], df_job.loc[i+1, 'Start'])

folder_path = './results/csv/vars/'
filenames = os.listdir(folder_path)
for filename in filenames:
    path = os.path.join(folder_path, filename)
    df = pd.read_csv(path, sep=';')

    # Check if there is a job unassigned
    pass # TODO
```
